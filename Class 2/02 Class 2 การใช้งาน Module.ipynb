# Module ไฟล์รวมคำสั่ง Python

## Global Variable ตัวแปรที่กำหนดไว้

Module ก็เป็นเช่นเดียวกับการใช้คำสั่งทั่วไป คือ มีการส่งต่อข้อมูลระหว่างคำสั่ง และเมื่อเรามีการส่งต่อข้อมูล จะต้องมีตัวแปรมาบันทึกข้อมูลนั้นไว้เพื่อใช้ส่งต่อ จริงๆ แล้วคำว่า Global Variable ประกอบไปด้วย 2 อย่าง คือ 1) Variable หรือตัวแปร 2) Global

Global เป็นเพียงชื่อเรียกตัวแปรที่มีการสร้าง และมี scope ที่สามารถเรียกใช้ได้ไม่ว่าจะทำงานกับส่วนไหนของไฟล์ หรือการนำเข้ามาทำงานกับอีกไฟล์เช่นที่เรากำลังจะดึง Global Variable จาก Module file_mangement มาใช้บน Jupyter Notebook
