# Pandas

การแก้ไขข้อมูล การเพิ่มข้อมูล ให้ระบุ column, row ที่ต้องการแก้ หรือ ที่ไม่มีอยู่เพื่อให้สร้างใหม่

## นำเข้าเครื่องมือ

```python
import pandas as pd
```

## ข้อมูลที่ใช้

ไฟล์ object ที่เปิดด้วย pandas จะเรียกว่า DataFrame

```python
east = pd.read_excel('excel/East.xlsx').set_index('Product')
new_column = east['Qtr1'] + east['Qtr2']
```

## เพิ่ม column ใหม่

```python
east['first half'] = new_column
```

```python
east
```

```python
east[['first half','Qtr1','Qtr2','Qtr3','Qtr4']]
```

## เพิ่ม row ใหม่

```python
cardamom = east.loc['Cardamom Seed (Ground)'] + east.loc['Cardamom Seed (Whole)']
```

```python
east.loc['cardamom'] = cardamom
```

```python
east.loc['Basil Leaf (Whole)'] = cardamom
```

```python
east['Qtr1']['Basil Leaf (Whole)'] = 9999
```
