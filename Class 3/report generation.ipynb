```python
def get_type(file):
    return os.path.splitext(file)[-1]
```

```python
def get_data(file_list,target_folder):
    data_list = []
    for file in file_list:
        file_type = get_type(file)
        file_path = os.path.join(target_folder,file)
        if file_type == '.csv':
            data_list.append(pd.read_csv(file_path).set_index('Product'))
        elif file_type == '.xlsx':
            data_list.append(pd.read_excel(file_path).set_index('Product'))
    return data_list
```

```python
data_list = get_data(file_list,target_folder)
data_list
```

```python
def sum_all(data_list):
    sum_result = pd.DataFrame()
    for data in data_list:
        sum_result = sum_result.add(data,fill_value=0)
    return sum_result
```

```python
sum_result = sum_all(data_list)
```

```python
def column_sum(sum_result):
    sum_result['Total Year'] = sum_result.sum(axis=1)
    sum_result.loc['Total Product'] = sum_result.sum(axis=0)
```
